# Visualization with Matplotlib

```python
import matplotlib.pyplot as plt #import plotting framework of matplotlib
%matplotlib inline
```

```python
l = [1,2,4,5,7,10,8]
l
```

```python
plt.plot(l) #create a line plot of l
plt.show() #show plot
```

```python
plt.plot(range(4,11), l) #defines x-axis values for y-axis values l
plt.show()
```

```python
plt.scatter(x = range(7), y = l) #create scatterplot
plt.show()
```
